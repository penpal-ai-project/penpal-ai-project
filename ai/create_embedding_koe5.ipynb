```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("nlpai-lab/KoE5")
print("모델 로드 완료")
```

```python
# KoE5 임베딩 벡터 생성 및 사용자별 저장
import torch


# 사용자별 임베딩 저장 딕셔너리
user_embeddings = {}


def add_letter(user_id, new_text):

    # KoE5 입력 형식
    text = "query: " + new_text

    # 문장을 임베딩 벡터로 변환
    embedding = model.encode(text, convert_to_tensor=True)

    # 사용자별 저장
    user_embeddings[user_id] = embedding

    print("현재 저장된 사용자 수:", len(user_embeddings))
    print(user_embeddings)


while True:

    user_id = input("\n사용자 ID 입력 (q 입력 시 종료): ")

    if user_id == "q":
        break

    text = input("편지 입력: ")

    add_letter(user_id, text)



# -------------------------------
# 이후 백엔드에서 이렇게 사용 가능
# -------------------------------

#all_embeddings = list(user_embeddings.values())

# print("\n전체 임베딩 개수:") => 확인용
# print(len(all_embeddings))

```
